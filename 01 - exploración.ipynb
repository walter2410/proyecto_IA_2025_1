```python
import os
import pandas as pd
import zipfile
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
os.environ['KAGGLE_CONFIG_DIR'] = os.path.expanduser('~/.kaggle')
```

```python
api = KaggleApi()
api.authenticate() 
```

```python
api.competition_download_files(
    'udea-ai-4-eng-20251-pruebas-saber-pro-colombia',
    path = 'archivos_proyecto',
    force = True, 
)
```

```python
carpeta = 'archivos_proyecto/'
```

```python
with zipfile.ZipFile(carpeta + 'udea-ai-4-eng-20251-pruebas-saber-pro-colombia.zip','r') as zip_ref:
    zip_ref.extractall(carpeta)
    
    
```

```python
for archivo in os.listdir(carpeta):
    print(archivo)
    
```

```python
df_train = pd.read_csv(carpeta + 'train.csv')
```

```python
df_train.head(9)
```

```python
df_train.iloc[8, 2]
```

```python
print('Estudiantes por departamento')
print(df_train['ESTU_PRGM_DEPARTAMENTO'].value_counts().head())
print()
print('Estudiantes estrato')
print(df_train['FAMI_ESTRATOVIVIENDA'].value_counts().head())
```

```python
print('Distribución de estudiantes en departamento según estrato')
agrupamiento = df_train.groupby(['ESTU_PRGM_DEPARTAMENTO','FAMI_ESTRATOVIVIENDA']).size().unstack(fill_value = 'walter').head()
agrupamiento.head(20)


```

```python
agrupamiento_1 = df_train.groupby(['ESTU_PRGM_DEPARTAMENTO','FAMI_ESTRATOVIVIENDA'])['RENDIMIENTO_GLOBAL'].value_counts().unstack(fill_value = 'ostias')
agrupamiento_1.head(20)
```

```python
filtro = df_train[df_train['ESTU_PRGM_DEPARTAMENTO'].isin(['ANTIOQUIA','BOGOTÁ'])]
filtro
```

```python
agrupamiento_2 = filtro.groupby(['ESTU_PRGM_DEPARTAMENTO','FAMI_ESTRATOVIVIENDA'])['RENDIMIENTO_GLOBAL'].value_counts().unstack()
agrupamiento_2.head(24)
```
